5. Limpieza y transformación de bank-additional

- Nulos 

```python
df_bank_additional.stb.missing()
```

```python
# Comprobación del % de nulos por columna (de mayor a menor)
(df_bank_additional.isna().mean() * 100).sort_values(ascending=False).head(15)
```

```python
df_bank_additional["age_missing"] = df_bank_additional["age"].isna().astype(int)
df_bank_additional["euribor_missing"] = df_bank_additional["euribor3m"].isna().astype(int)
```

```python
cols_cat_missing = ["default", "loan", "housing"]

for col in cols_cat_missing:
    df_bank_additional[col] = df_bank_additional[col].fillna("unknown")

# Las variables categóricas (`default`, `loan`, `housing`) presentan valores
# faltantes que se sustituyen por la categoría `"unknown"`, ya que representan
# información no disponible.
```

- Duplicados 

```python
# Duplicados totales y duplicados por id (si existe)
df_bank_additional.duplicated().sum(), df_bank_additional["id_"].duplicated().sum()

# No tenemos duplicados
```

- Columnas sobrantes

```python
# quitar unnamed
df_bank_additional = df_bank_additional.drop(columns=[c for c in df_bank_additional.columns if c.lower().startswith("unnamed")], errors="ignore")
df_bank_additional.shape
```

- Conversión de datos 

```python
# Convertir columnas con coma decimal a numero
num_comma_cols = ["emp.var.rate", "cons.price.idx", "cons.conf.idx", "euribor3m", "nr.employed"]

for col in num_comma_cols:
    if col in df_bank_additional.columns:
        df_bank_additional[col] = (
            df_bank_additional[col].astype("string")
            .str.replace(",", ".", regex=False)
        )
        df_bank_additional[col] = pd.to_numeric(df_bank_additional[col], errors="coerce")

df_bank_additional[num_comma_cols].dtypes
```

```python
meses = {
    "enero": "01",
    "febrero": "02",
    "marzo": "03",
    "abril": "04",
    "mayo": "05",
    "junio": "06",
    "julio": "07",
    "agosto": "08",
    "septiembre": "09",
    "octubre": "10",
    "noviembre": "11",
    "diciembre": "12"
}
for mes, num in meses.items():
    df_bank_additional["date"] = df_bank_additional["date"].str.replace(mes, num)
```

```python
df_bank_additional["date"] = pd.to_datetime(df_bank_additional["date"], dayfirst=True)
df_bank_additional["date"].head()
```

```python
df_bank_additional["date"].dtype
```

```python
# Limpiar columnas de categóricas
cat_cols = ["job", "marital", "education", "contact", "poutcome"]
# Quitamos espacios y minúsculas 
for col in cat_cols:
    df_bank_additional[col] = (
        df_bank_additional[col]
        .astype(str)
        .str.strip()      
        .str.lower()    
    )
```

```python
# Sultición de los missing values en las columnas categóricas por "unknown"
cat_cols = ["job", "marital", "education", "contact", "poutcome"]
for col in cat_cols:
    df_bank_additional[col]= df_bank_additional[col].fillna("unknown")
```

```python
# Comprobación de la sustituación 
df_bank_additional[cat_cols].isna().sum()
```

```python
# Limpieza de la columna "y"
df_bank_additional["y"] = (
    df_bank_additional["y"]
    .astype(str)
    .str.strip()
    .str.lower()
)
```

```python
# Sustitución de los valores de la columna "y" a binarios para un mejor análisis posterior 
df_bank_additional["y"] = df_bank_additional["y"].map({"yes": 1, "no": 0})
```

```python
df_bank_additional["y"].value_counts()
```

```python
df_bank_additional["cons.price.idx"] = df_bank_additional["cons.price.idx"].fillna(
    df_bank_additional["cons.price.idx"].median()
)

# La variable numérica `cons.price.idx` se imputa con la mediana para evitar
# perder registros y mantener la distribución de los datos.
```

- Imputar valores numericos 

```python
# Para "age"
df_bank_additional["age"] = df_bank_additional["age"].fillna(df_bank_additional["age"].median())
```

```python
# Para "euribor3m"
df_bank_additional["euribor3m"] = df_bank_additional["euribor3m"].fillna(df_bank_additional["euribor3m"].median())
```

- Crear features de fecha

```python
df_bank_additional["contact_year"] = df_bank_additional["date"].dt.year
df_bank_additional["contact_month"] = df_bank_additional["date"].dt.month
```

- Eliminación de fechas nulas 


```python

# Comprobación de cuántas fechas nulas hay
df_bank_additional["date"].isna().sum()
```

```python
# Eliminación de fechas
df_bank_additional = df_bank_additional.dropna(subset=["date"])
```

```python
# Comprobación 
df_bank_additional["date"].isna().sum()
```

```python
# Crear día de la semana del contacto
df_bank_additional["contact_weekday"] = df_bank_additional["date"].dt.day_name()
```

- Creación de la variable "was_contacted_before"

```python
df_bank_additional["was_contacted_before"] = (
    df_bank_additional["previous"]> 0
).astype(int)
```

```python
# Comprobación 
df_bank_additional["was_contacted_before"].value_counts()
```

```python
# Revisión de duplicados por id
df_bank_additional["id_"].duplicated().sum()
```

- Tratar "pdays" (999 significa que no se ha contactado)

```python
df_bank_additional["pdays"].value_counts()
# Vemos que tenemos 41175 clientes con los que no hemos contactado 
```

```python
# Sustituir 999 por "NaN"
df_bank_additional["pdays"] = df_bank_additional["pdays"].replace(999, "NaN")
```

```python
# Comprobación
df_bank_additional["pdays"].describe()
```

7. Limpieza y transformación de customer-details

- nulos

```python
df_customer_datails.stb.missing()
```

- Columnas inmencesarias 


```python
# quitar unnamed
df_customer_datails = df_customer_datails.drop(columns=[c for c in df_customer_datails.columns if c.lower().startswith("unnamed")], errors="ignore")
df_customer_datails.shape
```

- Duplicados

```python
# Duplicados 
df_customer_datails["ID"].duplicated().sum()

# No hay dumplicados 
```

- Transformación 

```python
# Revisión de la distribución de income
df_customer_datails["Income"].describe()

# Para ver en que rango de valores nos encontramos 
```

```python
# Valores de los hijos del cliente
df_customer_datails["Kidhome"].value_counts()
```

```python
df_customer_datails["Teenhome"].value_counts()
```

```python
# Crear feature de tamaño de familia
df_customer_datails["Family_size"]= (
    df_customer_datails["Kidhome"] +
    df_customer_datails["Teenhome"]
)

# Para saber cuántos hijos tiene cada cliente
```

```python
# Creamos features binarias para el posterior análisis 1 = tiene hijos, 0 = a no tiene hijos 
df_customer_datails["Has_kid"] = (df_customer_datails["Kidhome"]> 0 ).astype(int)
```

```python
df_customer_datails["Has_teens"] = (df_customer_datails["Teenhome"]> 0 ).astype(int)
```

```python
# Extraemos información de la fecha del cliente
# DT_Customer es la fecha en la que el cliente se registró

# Año de registro 
df_customer_datails["Customer_year"] = (df_customer_datails["Dt_Customer"]).dt.year
```

```python
# Mes de registro
df_customer_datails["Customer_month"] = (df_customer_datails["Dt_Customer"]).dt.month
```

```python
# Validar rango de fechas
df_customer_datails["Dt_Customer"].min(), df_customer_datails["Dt_Customer"].max()
```

```python
# Revisar visitas web
df_customer_datails["NumWebVisitsMonth"].describe()
```

```python
# Creación de la variable de antigüedad del cliente por años 
df_customer_datails["Customer_tenure_years"] = ((
    pd.Timestamp.today() - df_customer_datails["Dt_Customer"]
).dt.days / 365
).round(0)
```

8. Guardado del dataset limpio

```python
df_bank_additional["id_"].isin(df_customer_datails["ID"]).mean()
```

```python
# Merge de los dataset por id 
df_bank_marketing = df_bank_additional.merge(
    df_customer_datails, 
    left_on="id_", 
    right_on= "ID",
    how= "left"
)
```

```python
df_bank_marketing = df_bank_marketing.drop(columns=["ID"])
```

Tras realizar el merge entre `bank_additional` y `customer_details`,
se elimina la columna `ID` duplicada para mantener un único identificador (`id_`).

```python
# Creación de la variable "tenure_at_contact"
# Para saber cuánto tiempo llevaba el cliente cuándo recibió la llamada 
df_bank_marketing["tenure_at_contact_days"] = ((
    df_bank_marketing["date"] - df_bank_marketing["Dt_Customer"]
).dt.days /365 
).round(0)
```
